## Reduce Data Dimensionality With Principal Component Analysis

### Principal Component Analysis

An unsupervised, non-parametric statistical technique which primary use is to reduce dimensionality in machine learning.

REpresents data with a smaller subset of principal components.

makes maximum variability in dataset more visible by rotating the axes.

identifies a list of the principal axes to describe dataset.



Used to reduce feature set; this helps to :

** boost learning rates

** reduce computing cost

** more efficient models

#### 1st Principal Component in PCA:

** expresses most variance

Each additional component has less variance and more noise.

### Scaling the Data

```python
from sklearn.preprocessing import StandardScaler

standard_scaler = StandardScaler()
```

```python
X_scaled = standard_scaler.fit_transform(X)
X_scaled
```

Import PCA

```python
from sklearn.decomposition import PCA

NUMBER_OF_PRINCIPAL_COMPONENTS = 3

principal_component_analysis = PCA(n_components= NUMBER_OF_PRINCIPAL_COMPONENTS)

X_pca = principal_component_analysis.fit_transform(X_scaled)

X_pca

```

### Transform to a DataFrame

```python
X_pca_dataframe = pd.DataFrame(data=X_pca,
                columns=["Principal Component 1",
                         "Principal Component 2",
                         "Principal Component 3"],
                         index=X.index
                )
X_pca_dataframe
```

## find Clusters Using K_Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
MINIMUM_K = 1

MAXIMUM_K = 10

k = list(range(MINIMUM_K, MAXIMUM_K + 1))

inertia = []

for k_value in k:
    model = KMeans(n_clusters=k_value, random_state= 0)
    model.fit(X_pca_dataframe)
    inertia.append(model.inertia_)

inertia
```

```python
elbow_data = {"k": k,
              "inertia":inertia}

elbow_data_df = pd.DataFrame(elbow_data)
elbow_data_df
```

### Plotting the  elbow graph

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,15))

plt.plot(k, inertia, "bx-")

plt.xlabel("K (Numbers of Clusters")

plt.ylabel("Distortion (inertia)")
plt .show()
```

### Performing MAchine with the Optimal Number of Clusters

```python
OPTIMAL_K = 4

KMeans(n_clusters=OPTIMAL_K, random_state= 0)

model.fit(X_pca_dataframe)

predictions = model.predict(X_pca_dataframe)

```

A new dataFrame that shows the predicted clusters and the cryptocurrency features

```python
X_pca_dataframe["Clusters"] = model.labels_
X_pca_dataframe
```

```python
#join dataframes

final_df = pd.concat([df_with_coin_names,
                     X_pca_dataframe], axis = COLUMN_INDEX)

final_df
```

### Visualize Clusters 

```python
import plotly.express as express

figure = express.scatter_3d(
            final_df,
            x = "Principal Component 1",
            y = "Principal Component 2",
            z = "Principal Component 3",
            color = "Clusters",
            hover_name= "CoinName",
            width= 1000

)

figure.update_layout(legend = dict(x= 0, y= 1))

figure.show()
```
